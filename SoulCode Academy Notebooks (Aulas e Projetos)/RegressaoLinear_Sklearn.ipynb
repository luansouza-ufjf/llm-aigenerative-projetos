```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
#Regression
x,y = make_regression(n_samples=400, n_features=1, noise=10)

plt.scatter(x,y)
```

image.png

```python
modelo = LinearRegression()
modelo.fit(x,y)
```

```python
a_coeff = modelo.coef_
l_coeff = modelo.intercept_

#y = ax + b
#y = a_coeff * x + l_coeff
```

```python
print('Coeficiente Angular: {0}\nCoeficiente Linear {1}'.format(a_coeff,l_coeff))
```

```python
#Valor de Y para X = 1.5
l_coeff + a_coeff*1.9
```

```python
plt.scatter(x,y, label = "Dados Histricos")
plt.plot(x, a_coeff*x + l_coeff , color='red', label ="Linha de Predio")
plt.scatter(1.5, l_coeff + a_coeff*1.5, color='green')
plt.legend()
plt.show()
```
