```python
import pandas as pd
```

```python
df = pd.read_csv('Dataset2.csv')
df.info()
```

```python
df.replace("", pd.NA, inplace=True)
```

```python
df.head(100)
```

```python
id_vars = ["DATASET", "SERIES_CODE", "OBS_MEASURE", "COUNTRY.Name",
           "INDEX_TYPE.Name", "COICOP_1999.Name", "TYPE_OF_TRANSFORMATION.Name",
           "FREQUENCY.Name", "SCALE.Name"]

df_long = df.melt(
    id_vars=id_vars,
    var_name="Date",
    value_name="Value"
)
```

```python
df_long[["Year", "Month"]] = df_long["Date"].str.split("-M", expand=True)
df_long["Month"] = df_long["Month"].fillna("00")  # Para dados anuais
df_long["Date"] = pd.to_datetime(df_long["Year"] + "-" + df_long["Month"], format="%Y-%m", errors="coerce")
```

```python
df_netherlands = df_long[
    (df_long["COUNTRY.Name"] == "Netherlands, The") &
    (df_long["FREQUENCY.Name"] == "Monthly")
]
```
