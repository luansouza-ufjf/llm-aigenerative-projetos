```python
'''1.	Uma pesquisa de mercado revelou que 240 dentre 600 pessoas de uma localidade, escolhidas aleatoriamente, comprariam certo produto se o mesmo for lançado no mercado da localidade.
Construa intervalos de confiança de 95% e 99% para a proporção de pessoas da localidade que comprariam o referido produto.
'''
```

```python
n = 600
x = 240
p = x / n
confiancas = [0.95, 0.99]

for conf in confiancas:
    z = sp.norm.ppf((1 + conf) / 2)
    margin = z * (p * (1 - p) / n)**0.5
    print(f"Intervalo de Confiança {conf*100}%: ({p - margin:.4f}, {p + margin:.4f})")
```

```python
'''2.	O fabricante de certa marca de suco informa que as embalagens de seu produto têm em média 500 ml, com desvio padrão igual a 10 ml.
Tendo sido encontradas no mercado algumas embalagens com menos de 500 ml, suspeita-se que a informação do fabricante seja falsa.
Para verificar se isto ocorre, um fiscal analisa uma amostra de 200 embalagens escolhidas aleatoriamente no mercado e constata que as mesmas contêm em média 498 ml.
Considerando-se um nível de significância de 5%, pode-se afirmar que o fabricante está correto?
Calcule o valor da prova para esta amostra.'''
```

```python
import scipy.stats as sp

mu = 500
desvio = 10
n = 200
media = 498
alpha = 0.05

z = (media - mu) / (desvio / (n**0.5))
p_valor = 2 * (1 - sp.norm.cdf(abs(z)))
print(f"Valor da Prova: {z}")
print(f"P-Valor: {p_valor}")

z_critico = sp.norm.ppf(1 - alpha / 2)
print(f"Valor crítico: ±{z_critico:.4f}")

# Teste unilateral à esquerda (H1: mu < 500)
z_critico_negativo = sp.norm.ppf(alpha)
ztest = (media - mu) / (desvio / (n**0.5))
print(f"Valor do teste Z: {ztest:.4f}")

# Decisão via região crítica
if ztest < z_critico_negativo:
    print('Rejeitamos H0')
else:
    print('Não Rejeitamos H0')

# Decisão via p-valor
p_valor = sp.norm.cdf(ztest)
print(f'P-Valor é {p_valor:.8f}')
if p_valor < alpha:
    print('Rejeitamos H0')
else:
    print('Não Rejeitamos H0')
```

```python
'''3.  	A duração das lâmpadas produzidas por certo fabricante tem distribuição normal com média igual a 1200 horas e desvio padrão igual a 300 horas.
O fabricante introduz um novo processo na produção das lâmpadas.
Para verificar se o novo processo produz lâmpadas de maior duração, o fabricante observa 100 lâmpadas produzidas pelo novo processo e constata que as mesmas duram em média 1265 horas.
Admitindo-se um nível de significância de 5%, pode-se concluir que o novo processo produz lâmpadas com maior duração?
'''
```

```python
mu = 1200
desvio = 300
n = 100
media = 1265
alpha = 0.05

# Teste unilateral à direita (H1: mu > 1200)
z_critico_positivo = sp.norm.ppf(1 - alpha)
ztest = (media - mu) / (desvio / (n**0.5))
print(f"Valor do teste Z: {ztest:.4f}")

# Decisão via região crítica
if ztest > z_critico_positivo:
    print('Rejeitamos via H0')
else:
    print('Não Rejeitamos via H0')

# Decisão via p-valor
p_valor = 1 - sp.norm.cdf(ztest)
print(f'P-Valor é {p_valor:.8f}')
if p_valor < alpha:
    print('Rejeitamos via H0')
else:
    print('Não Rejeitamos via H0')
```

```python
mu = 57
n = 100
media = 56.91
sigma = 0.24
alpha = 0.05

# Teste bilateral (H1: mu ≠ 57)
t_critico_positivo = abs(sp.t.ppf(alpha/2, df=n-1))
ztest = (media - mu) / (sigma / (n**0.5))
print(f"Valor do teste Z: {ztest:.4f}")
# Decisão via região crítica
if ttest < -t_critico_positivo or ttest > t_critico_positivo:
    print('Rejeitamos H0')
else:
    print('Não Rejeitamos H0')

# Decisão via p-valor
p_valor = 2 * (1 - sp.t.cdf(abs(ztest), df=n-1))
print(f'P-Valor é {p_valor:.8f}')
if p_valor < alpha:
    print('Rejeitamos H0')
else:
    print('Não Rejeitamos H0')

print('Portanto concluo que o processo está sob controle')
```

```python
''' 6.) Numa localidade, 32% dos consumidores consomem determinado produto.
Foi lançado no mercado da localidade um produto concorrente.
Uma pesquisa realizada com 500 consumidores escolhidos ao acaso revelou que 145 dentre estes consomem o antigo produto.
Pode-se concluir, num nível de significância de 2%, que a preferência pelo produto antigo diminuiu com a entrada do concorrente no mercado?
Calcule o valor da prova para esta amostra.'''
```

```python
p0 = 0.32
n = 500
x = 145
alpha = 0.02

# Teste unilateral à esquerda (H1: p < 0.32)
p_amostral = x / n
z_critico_negativo = sp.norm.ppf(alpha)
ztest = (p_amostral - p0) / ((p0 * (1 - p0) / n)**0.5)
print(f"Valor do teste Z: {ztest:.4f}")


# Decisão via região crítica
if ztest < z_critico_negativo:
    print('Rejeitamos H0')
else:
    print('Não Rejeitamos H0')

# Decisão via p-valor
p_valor = sp.norm.cdf(ztest)
print(f'P-Valor é {p_valor:.8f}')
if p_valor < alpha:
    print('Rejeitamos H0')
else:
    print('Não Rejeitamos H0')
```

```python
'''7.) Sabe-se que 6% das unidades de certo produto são substituídas gratuitamente por apresentar defeitos de fabricação.
Para reduzir este percentual, o fabricante investiu na melhoria da qualidade do produto.
Consta-se que 12 dentre 400 unidades vendidas tiveram que ser substituídas gratuitamente por apresentar defeitos de fabricação.
 Pode-se concluir, num nível de significância de 3%, que a qualidade do produto melhorou?'''
```

```python
p0 = 0.06
n = 400
x = 12
alpha = 0.03

# Teste unilateral à esquerda (H1: p < 0.06)
p_amostral = x / n
z_critico_negativo = sp.norm.ppf(alpha)
ztest = (p_amostral - p0) / ((p0 * (1 - p0) / n)**0.5)
print(f"Valor do teste Z: {ztest:.4f}")

# Decisão via região crítica
if ztest < z_critico_negativo:
    print('Rejeitamos H0')
else:
    print('Não Rejeitamos H0')

# Decisão via p-valor
p_valor = sp.norm.cdf(ztest)
print(f'P-Valor é {p_valor:.8f}')
if p_valor < alpha:
    print('Rejeitamos H0')
else:
    print('Não Rejeitamos H0')

print('Portanto concluo que a qualidade do produto melhorou')
```
