```python
Angola = df['Angola']
Uganda = df['Uganda']
Nigeria = df['Nigéria']
Lithuania = df['Lituânia']
Netherlands = df['Netherlands']
differences = []
countries = ['Angola:', 'Uganda:', 'Nigeria:', 'Lithuania:', 'Netherlands:']
means = [Angola.mean(), Uganda.mean(), Nigeria.mean(), Lithuania.mean(), Netherlands.mean()]

for i in range(len(countries)):
  for j in range(i + 1, len(countries)):
    difference = abs(means[i] - means[j])
    differences.append({'País 1': countries[i], 'País 2': countries[j], 'Diferença/Médias': difference})

df_differences = pd.DataFrame(differences)
df_differences
```

```python
# prompt: Aplicar o teste t entre Angola e Netherlands

from scipy import stats

# Aplicar o teste t para amostras independentes
t_stat, p_value = stats.ttest_ind(df["Uganda"], df["Netherlands"])

# Exibir os resultados
print(f"Estatística t: {t_stat}")
print(f"Valor p: {p_value}")

# Interpretação
alpha = 0.05  # Nível de significância
if p_value < alpha:
    print("Diferença estatisticamente significativa entre os grupos.")
else:
    print("Não há diferença estatisticamente significativa entre os grupos.")
```
