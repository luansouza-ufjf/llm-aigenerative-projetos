Distribuição por Escolaridade e Salário

```python
df_departamento = df.groupby('departamento')['salario'].mean().reset_index()
df_departamento = df_departamento.sort_values(by='salario', ascending=False)
```

```python
df
```

```python
df_idade = df.groupby('idade')['salario'].mean().reset_index()
df_idade = df_idade.sort_values(by='salario', ascending=False)
```

```python
df_estado = df.groupby('estado')['salario'].mean().reset_index()
df_estado = df_estado.sort_values(by='salario', ascending=False)
```

```python
df
```

```python
from matplotlib import pyplot as plt
df['idade'].plot(kind='hist', bins=20, title='idade')
plt.title('Distribuição por idade')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
plt.bar(df['escolaridade'],df['salario'])
plt.title('Distibuição de Salário por Escolaridade')
plt.xlabel('Escolaridade')
plt.ylabel('Salário')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.barh(df_estado['estado'], df_estado['salario'])
plt.title('Qual Estado paga Melhor')
plt.xlabel('Salário')
plt.ylabel('Estado')
plt.show()
```

Média Salarial por Departamento

```python
plt.figure(figsize=(8, 8))
plt.pie(df_departamento['salario'], labels=df_departamento['departamento'], autopct='%1.1f%%', startangle=90)
plt.title('Média Salarial por Departamento')
plt.axis('equal')
plt.show()
```

Agrupamentor por Nome e Nota
