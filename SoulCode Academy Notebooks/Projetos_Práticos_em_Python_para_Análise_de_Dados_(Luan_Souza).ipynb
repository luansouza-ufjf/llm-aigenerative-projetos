```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
Base_Dados = pd.read_csv('Startups 2022.csv')

Base_Dados.shape
```

```python
#Primeiros Registros
Base_Dados.head()
```

```python
#Colunas Existentes
Base_Dados.columns
```

```python
#Renomear para ficar melhor identificação
Base_Dados.rename(columns= {
    'Unnamed: 0': 'Id',
    'Company': 'Empresa',
    'Valuation($B)': 'Valor($)',
    'Data Joined': 'Data de Adesão',
    'Country': 'País',
    'City': 'Cidade',
    'Industry': 'Setor',
    'Select Investors': 'Investidores'
}, inplace=True)

```

```python
#Verificando a Base de Dados
Base_Dados.info()
```

```python
# Campos Nulos
Base_Dados.isnull().sum()
```

```python
#Visualização Gráfica
plt.figure(figsize=(15,6))
plt.title('Campos Nulos')
sns.heatmap(Base_Dados.isnull())
```

```python
#Campos Únicos

Base_Dados.nunique()
```

```python
Analise = round(Base_Dados['País'].value_counts(normalize=True)*100,1)
```

```python
plt.figure( figsize=(15,6) )
plt.title('Campos Únicos')
plt.pie(
    Analise,
    labels = Analise.index,
    shadow = True,
    startangle=90,
    autopct='%1.1f%%'

);
```
